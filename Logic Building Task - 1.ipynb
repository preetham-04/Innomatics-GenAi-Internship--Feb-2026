2) Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
Pass_count = 0
fail_count = 0
for elements in marks:
    #count using iteration
    if elements >= 50:
        Pass_count += 1
    else:
        fail_count += 1
print("Students pass count:", Pass_count)
print("Students Fail count:", fail_count)
```

3) Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
names_list = []
for word in  names:
    # Remove extra spaces and convert the characters to lowercase
    w=word.strip()
    names_list.append(w.lower())

print("Standered names list :",names_list)
```

4) Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    size = len(message) # for length of message
    print("Message :",message," / Length :",size)
    if size > 10:
        print("This message is long")
```

5) Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
errors_count = 0
for entry in logs:
    if entry == "ERROR":
        errors_count += 1
print("Number of ERROR entries in logs :", errors_count)
```
